```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
x = np.random.normal(scale=2, size=50)
y = np.random.normal(scale=2, size=50)
```

```python
plt.figure(figsize=(10, 10), dpi=100)
plt.grid(False)
plt.axis('off')
color = sns.color_palette("Paired")
cmap = sns.cubehelix_palette(start=2, light=1, hue=1, as_cmap=True)
ax = sns.kdeplot(x=x, y=y, fill=True, levels=50, cmap=cmap, thresh=0)
ax.text(-7, -8.5, "KINDNESS", 
       fontsize = 40,
       fontname = 'Montserrat',
       weight = 900,
       color='#333333')  
ax.text(-7, -9, "Be kind in this cruel world", 
       fontsize = 8,
       fontname = 'Montserrat',
       color='#333333') 
ax.text(-7, 0, "o", 
       fontsize = 8,
       fontname = 'Montserrat',
       color='#cccccc') 
ax.text(-7, 9, "1 / 3", 
       fontsize = 8,
       fontname = 'Montserrat',
       color='#333333') 
ax.text(7, -6.5, "Sometimes  it  takes  only  one  act  of  kindness  and  caring  to  change  a  person's  life. - Jackie  Chan", 
       fontsize = 8,
       fontname = 'Montserrat',
       color='#cccccc',
       rotation=90) 
plt.savefig("poster-1-kindness.png", format="png", bbox_inches="tight", dpi=100)
```

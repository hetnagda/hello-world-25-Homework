```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
x = np.random.normal(scale=2, size=50)
y = np.random.normal(scale=2, size=50)
```

```python
plt.figure(figsize=(10, 10), dpi=100)
plt.sharex=True
plt.sharey=True
plt.grid(False)
ax = plt.axes([0., 0., 1., 1.], frameon=False, xticks=[],yticks=[])
plt.axis('off')
cmap = sns.cubehelix_palette(start=2.5, light=1, hue=1, as_cmap=True)
ax = sns.kdeplot(x=x, y=y, fill=True, levels=50, cmap=cmap, thresh=0)
ax.text(.02, .05, "FRIENDSHIP", 
       fontsize = 40,
       fontname = 'Montserrat',
       weight = 900,
       color='#333333',transform=ax.transAxes)  
ax.text(.02, .03, "Friends are true gems", 
       fontsize = 8,
       fontname = 'Montserrat',
       color='#333333',transform=ax.transAxes) 
ax.text(.02, .4, "o", 
       fontsize = 8,
       fontname = 'Montserrat',
       color='#cccccc',transform=ax.transAxes) 
ax.text(.02, .97, "3 / 3", 
       fontsize = 8,
       fontname = 'Montserrat',
       color='#333333',transform=ax.transAxes) 
ax.text(.98, .25, "True friends are like diamonds - bright, beautiful, valuable, and always in style - Nicole Richie", 
       fontsize = 8,
       fontname = 'Montserrat',
       color='#cccccc',
       rotation=90,transform=ax.transAxes) 
plt.savefig("poster-3-friendship.png", format="png", bbox_inches="tight", dpi=100)
```

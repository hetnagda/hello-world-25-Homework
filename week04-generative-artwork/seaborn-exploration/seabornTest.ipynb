```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
tips = sns.load_dataset("tips")
```

```python
tips
```

```python
sns.set()
sns.histplot(x='total_bill', data=tips, color="red")
```

```python
sns.set()
sns.kdeplot(x='total_bill', data=tips, color="red")
```

```python
sns.set()
sns.scatterplot(x='tip',y='total_bill', data=tips, color="red")
```

```python
sns.set()
sns.kdeplot(x='tip',y='total_bill', data=tips, color="red")
```

```python
sns.set()
sns.kdeplot(x='tip', y='total_bill', data=tips, fill=True, n_levels=15)
```

```python
x = np.random.normal(scale=2, size=50)
y = np.random.normal(scale=2, size=50)
```

```python
plt.figure(figsize=(8,8))
plt.grid(False)
plt.axis('off')
color = sns.color_palette("Paired")
cmap = sns.cubehelix_palette(start=2, light=1, hue=1, as_cmap=True)
ax = sns.kdeplot(x=x, y=y, fill=True, levels=50,thresh=0, cmap=cmap)
ax.text(-7.5, -6.5, "KINDNESS", 
       fontsize = 32,
       fontname = 'Montserrat',
       weight = 900,
       color='#333333')  
ax.text(-7.4, -6.9, "Be kind in this cruel world", 
       fontsize = 8,
       fontname = 'Montserrat',
       color='#333333') 
ax.text(-7.4, 0, "o", 
       fontsize = 8,
       fontname = 'Montserrat',
       color='#333333') 
ax.text(-7.4, 6.9, "1 / 3", 
       fontsize = 8,
       fontname = 'Montserrat',
       color='#333333') 
ax.text(9, -6.5, "Sometimes  it  takes  only  one  act  of  kindness  and  caring  to  change  a  person's  life. - Jackie  Chan", 
       fontsize = 8,
       fontname = 'Montserrat',
       color='#cccccc',
       rotation=90) 
plt.savefig("myImagePDF.pdf", format="pdf", bbox_inches="tight")
```
